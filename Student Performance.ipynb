```python
sns.set_style('whitegrid')
```

```python
df=pd.read_csv("Students Performance.csv")
df
```

```python
sns.barplot(x="gender",y="math score",data=df,palette="plasma")
```

```python
sns.countplot(x="gender",data=df)
```
